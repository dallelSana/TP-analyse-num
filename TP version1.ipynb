### 1) * Déclaration de deux fonction p(x) et f(x)

```python
def p(x): 
    return (x-1)**6 
def f(x):
    return x**6 - 6*x**5 + 15*x**4 - 20*x**3 + 15*x**2 - 6*x + 1.

print('p({})={}'.format(0.998,p(0.998)))
print('f({})={}'.format(0.998,f(0.998)))
```

### Remarque: 
#### théoriquement ces deux fonctions sont identiques mais lorsqu'on choisit une valeur trés proche de 1 on trouve des résultats différents !!

## * Evaluation p(x) et f(x) pour x allant de .998 à 1.002 en choisissant des pas de .0001

```python
import numpy as np
print('Fonction réduite')
print('            ')
for i in np.arange(0.998, 1.002,0.0001): #Renvoie des valeurs régulièrement espacées. 
                                         #(start=0.998 , stop=1.002, step=0.0001)
    print('p(i)= ' ,p(i))#Remplacer x par la valeur de i puis le calculé dans la fonction
print('')
print('------------------------------------------')
print('')
print('Fonction developpée')
print('')
for i in np.arange(0.998, 1.002,0.0001):
    print('f(i)= ' ,f(i))
print('------------------------------------------')

print(p(i)==f(i)) #Pour démontrer que les deux fonctions ne donnent pa la même valeur
```

## Remarque:
### On trouve une erreur lors de l'utilisation des valeurs réels : Il y'a un problème d'arrondissment 

## 2) Présentation graphique:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.998, 1.002, 0.0001)
p = lambda x:(x-1)**6 #lambda prend à chaque fois la valeur de x et retourne la valeur de l'expression
fig = plt.figure(1, figsize=(15, 7))
plt.plot(x,p(x)) 
f = lambda x: x**6 - 6*x**5 + 15*x**4 - 20*x**3 + 15*x**2 - 6*x + 1
fig=plt.figure(1, figsize=(15 , 10))
plt.plot(x,f(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graphe1: Les fonction p(x) et f(x)')
plt.legend(('p(x)', 'f(x)'), loc='upper center', shadow=True)
plt.grid(True)   
plt.show()
```
